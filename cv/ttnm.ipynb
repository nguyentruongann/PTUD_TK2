```python
import pandas as pd
import matplotlib.pyplot as plt
# Tạo dữ liệu
data = {
    'Tiêu Chí Lựa Chọn': ['Giá cả', 'Giá cả', 'Chủng loại', 'Giá cả', 'Chủng loại', 'Kích thước', 'Kích thước', 'Chủng loại', 'Giá cả'],
    'Mức Giá': ['1 triệu – 2 triệu', '2 triệu – 3 triệu', '1 triệu – 2 triệu', '1 triệu hoặc dưới', '1 triệu hoặc dưới', '1 triệu – 2 triệu', '1 triệu – 2 triệu', '2 triệu – 3 triệu', '1 triệu – 2 triệu'],
    'Thời Gian Mua (phút)': [10, 20, 30, 10, 40, 20, 10, 10, 20],
    'Tính Năng Hài Lòng': [['A', 'B'], ['A', 'C'], ['A', 'B'], ['B', 'C'], ['A', 'B', 'D'], ['B', 'C'], ['A', 'D'], ['A', 'C'], ['A', 'B']]
}
df = pd.DataFrame(data)
# Tính toán thống kê cho thời gian mua sản phẩm
mean_time = df['Thời Gian Mua (phút)'].mean()
median_time = df['Thời Gian Mua (phút)'].median()
mode_time = df['Thời Gian Mua (phút)'].mode()[0]
# Chuẩn bị dữ liệu tính năng hài lòng
feature_satisfaction = {}
for features in df['Tính Năng Hài Lòng']:
    for feature in features:
        if feature in feature_satisfaction:
            feature_satisfaction[feature] += 1
        else:
            feature_satisfaction[feature] = 1
# result: {'A': 7, 'B': 6, 'C': 4, 'D': 2}
# Biểu đồ tính năng hài lòng
plt.figure(figsize=(10, 6))
plt.bar(feature_satisfaction.keys(), feature_satisfaction.values(), color='skyblue')
plt.title('Mức Độ Hài Lòng Với Tính Năng Sản Phẩm')
plt.xlabel('Tính Năng')
plt.ylabel('Số Lần Được Chọn')
plt.show()

mean_time, median_time, mode_time, feature_satisfaction
```

```python
# Dữ liệu thời gian lựa chọn và mua sản phẩm
time_to_purchase = [10, 20, 30, 10, 40, 20, 10, 10, 20]
# Biểu đồ
plt.figure(figsize=(8, 5))
plt.hist(time_to_purchase, bins=range(min(time_to_purchase), max(time_to_purchase) + 10, 10), color='skyblue', alpha=0.7)
plt.axvline(mean_time, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_time:.2f} min')
plt.axvline(median_time, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_time} min')
plt.axvline(mode_time, color='blue', linestyle='dashed', linewidth=1, label=f'Mode: {mode_time} min')
plt.title('Thời Gian Lựa Chọn và Mua Sản Phẩm')
plt.xlabel('Thời Gian (phút)')
plt.ylabel('Số Người Dùng')
plt.legend()
plt.show()
mean_time, median_time, mode_time
```
